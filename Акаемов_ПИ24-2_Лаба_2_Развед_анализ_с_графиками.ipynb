#### **Время-временные ряды**

Создадим отдельный df, состоящий только из дат(столбец date)

```python
dftime=df.loc[:,'date']
dftime
```

__Частота операций__

```python
dftime.index = pd.to_datetime(dftime.index, unit='d')
freq = pd.infer_freq(dftime.index)
diffs = dftime.index.to_series().diff()
print(f"Частота: {freq}, Интервал: {diffs.mean()}")
```

Наблюдаем, что операции производились ежедневно, при этом информация о конкретном времени проведения операций в данных отсутствует. Это означает, что все котировки представляют собой итоговые дневные значения - цены открытия, закрытия, максимумы и минимумы за каждые торговые сутки. Такая структура данных характерна для исторических финансовых временных рядов и позволяет анализировать общие рыночные тенденции, но не дает возможности изучать внутридневную волатильность. Отсутствие временных меток ограничивает анализ краткосрочных ценовых движений, однако вполне достаточно для долгосрочного статистического исследования рынка.

__Пропуски в датах__

```python
full_range = pd.date_range(dftime.index.min(), dftime.index.max(), freq=freq)
missing = len(full_range) - len(df)
missing
```

В данных нет пропусков - все показатели заполнены за каждый день наблюдений. Это означает, что у всех криптовалют ежедневно фиксировались цены и объемы торгов без пробелов в информации.

Вычислим максимальную длину серии операций(без пропусков дней)

```python
gaps = (diffs > diffs.mode()[0]).cumsum()
max_series = dftime.groupby(gaps).size().max()
print(f"Макс. серия: {max_series}")
```

Максимальная серия данных составляет 72 946 наблюдений, что означает непрерывный ежедневный сбор информации на протяжении примерно 200 лет. Однако этот показатель явно не соответствует реальности для крипторынка, который существует менее 15 лет. Вероятно, в данных присутствуют дублирующиеся записи или техническая ошибка в формировании временного ряда, что требует дополнительной проверки качества dataset.

__Использование методов обнаружения выбросов (Z-показатели, IQR-метод)__

```python
# 1. Z-показатели (стандартные отклонения)
z_scores = np.abs((df['close'] - df['close'].mean()) / df['close'].std())
outliers_z = df[z_scores > 3]
print(f"Выбросы по Z-показателю: {len(outliers_z)}")

# 2. IQR-метод
Q1 = df['close'].quantile(0.25)
Q3 = df['close'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers_iqr = df[(df['close'] < lower_bound) | (df['close'] > upper_bound)]
print(f"Выбросы по IQR: {len(outliers_iqr)}")
```

Метод Z-показателя (1017 выбросов) выявляет только самые экстремальные отклонения от среднего, обычно за пределами 3 стандартных отклонений. Это действительно аномальные значения.

Метод IQR (14 018 выбросов) является более чувствительным и показывает все значения за пределами 1.5 межквартильного размаха. Столь большое количество выбросов этим методом подтверждает, что крипторынок по своей природе чрезвычайно волатилен - резкие скачки цен являются не исключением, а скорее правилом.

Разница в количестве выбросов (в 14 раз) доказывает, что криптовалюты характеризуются постоянной высокой волатильностью, где значительные ценовые движения являются обычным явлением.
